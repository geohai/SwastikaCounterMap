```python
import pandas as pd
import geocoder
import datetime
import time
#for checking image link or valid link
import requests
import validators
from urlextract import URLExtract
headers = { 'User-Agent': 'Mozilla/5.0 (Windows NT 6.1; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/32.0.1700.107 Safari/537.36' }
#######################################

df = pd.read_csv("final_csv_data.csv")
df_loc = pd.read_csv("geocoded_cities.csv")
print(df)
print(df.columns)
```

```python
def is_link_valid(link):
    #validating may be overkill, but we want to make sure that every site is reachable from the front end!
    val = validators.url(link)
    r = ''
    if val:
        while r == '':
            try:
                r = requests.get(url=link, timeout=10, headers=headers)
                break
            except:
                break
    if r != '':
        if r.status_code != 404:
            return True
        return False
    return False
```

```python
extractor = URLExtract()

lon_list = []
lat_list = []
img_list = []
is_img_list = []
found_address_list = []

i = 0
for ind in df.index:
    i += 1
    
    ###################################################
    #Identify location, lat and long by city and state#
    ###################################################
    city = df['city'][ind]
    state = df['STATE'][ind]
    
    cityState = str(city) + "," + str(state)
    g = geocoder.bing(cityState, key='AnSJDRrLPJczP613_0oVzeYr4Uiai6BpbYMMjz3abpR1EJ7EAZ4qfv316QHW9JMv')
    if g.json != None:
        lon_list.append(g.json['lng'])
        lat_list.append(g.json['lat'])
        found_address_list.append(g.json['address'])
    else:
        lon_list.append(None)
        lat_list.append(None)
        found_address_list.append('Not Found')
    print(i ,' / ', len(df[['city']]))
    ###################################################
    #End Block ########################################
    ###################################################
    
    #finding image link! Process: extract urls from string, detect if link is valid, and detect 404
    img_link = str(df['url_to_jpg'][ind])
    is_img = False # predetermining false to eliminate possible empty fields. (Short-hand!)
    img = ''
    urls = extractor.find_urls(img_link)
    
    if len(urls) > 0:
        is_img = is_link_valid(urls[0])
        img = urls[0]
        print(urls[0])
        
    is_img_list.append(is_img)
    img_list.append(img)
    print('is it an image? ' + str(is_img))
        
    
```

```python
#from pandas_geojson import to_geojson
#from pandas_geojson import write_geojson
import pandas_geojson as pdg
new_df = new_df.fillna("none")
new_df.rename(columns={'img_available':'Image?'}, inplace=True)
new_df['coordinates'] = new_df.reset_index()[['longitude', 'latitude']].values.tolist()
new_df['geojson_type'] = 'point'
properties = new_df.drop(['latitude', 'longitude'], axis=1)
properties = properties.fillna("none")
new_df
geo_json = pdg.GeoJSON.from_dataframe(new_df
                                ,geometry_type_col='geojson_type'
                                , coordinate_col='coordinates'
                                , property_col_list=properties)

pdg.save_geojson(geo_json,'swastika_data_aug30.geojson',indent=4)
geo_json
```
